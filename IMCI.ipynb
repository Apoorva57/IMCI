Specialized function for brain tumor

```python
def brain_tumor_mri_predi(path_image):
    labels=['Glioma (Astrocitoma, Ganglioglioma, Glioblastoma, Oligodendroglioma, Ependimoma) T1','Glioma (Astrocitoma, Ganglioglioma, Glioblastoma, Oligodendroglioma, Ependimoma) T1C+','Glioma (Astrocitoma, Ganglioglioma, Glioblastoma, Oligodendroglioma, Ependimoma) T2','Meningioma (de Baixo Grau, Atípico, Anaplásico, Transicional) T1','Meningioma (de Baixo Grau, Atípico, Anaplásico, Transicional) T1C+','Meningioma (de Baixo Grau, Atípico, Anaplásico, Transicional) T2','NORMAL T1','NORMAL T2','Neurocitoma (Central - Intraventricular, Extraventricular) T1','Neurocitoma (Central - Intraventricular, Extraventricular) T1C+','Neurocitoma (Central - Intraventricular, Extraventricular) T2','Outros Tipos de Lesões (Abscessos, Cistos, Encefalopatias Diversas) T1','Outros Tipos de Lesões (Abscessos, Cistos, Encefalopatias Diversas) T1C+','Outros Tipos de Lesões (Abscessos, Cistos, Encefalopatias Diversas) T2','Schwannoma (Acustico, Vestibular - Trigeminal) T1','Schwannoma (Acustico, Vestibular - Trigeminal) T1C+','Schwannoma (Acustico, Vestibular - Trigeminal) T2']
    path_model="/content/tumor_detection_1.h5"
    resnet_model=tf.keras.models.load_model(path_model)
    img=tf.image.resize(path_image,(224,224))
    img_arr=tf.keras.utils.img_to_array(img)
    x = preprocess_input(np.expand_dims(img_arr.copy(), axis=0))
    prediction=resnet_model.predict(x)
    ret = {}
    for x,y in enumerate(labels):
      ret[y]=prediction[0][x]
    jojo='Prediction using specialized ml model that the patient has a specific type of brain tumor, accurate up to 90% \n '
    for x in ret:
      jojo+=f"{x} : {ret[x]}\n "
    jojo+="prediction ends here\n"
    f= open('results.yml','a')
    f.write(jojo)
    f.close()


    return ret, gr.update(visible=True)
```

```python
def alzheimer_mri_predi(path_image):
    from PIL import Image
    img_pil = Image.fromarray((path_image * 255).astype(np.uint8))
    # Use a pipeline as a high-level helper
    from transformers import pipeline
    pipe = pipeline("image-classification", model="andrewromitti/alzheimer_model_aug")
    x = pipe(img_pil)
    ou='Prediction using specialized ml models that the patient has alzheimer that are accurate 94% of times \n '
    for i in x:
      ou+=f"{i['label']} : {i['score']}\n "
    ou+="prediction ends here\n"
    f= open('results.yml','a')
    f.write(ou)
    f.close()
    print(ou)
    return x
```

```python
def skin_cancer_predi(path_image):
    from PIL import Image
    img_pil = Image.fromarray((path_image * 255).astype(np.uint8))
    # Use a pipeline as a high-level helper
    from transformers import pipeline
    pipe = pipeline("image-classification", model="Anwarkh1/Skin_Cancer-Image_Classification")
    x = pipe(img_pil)
    ou='Prediction of specialized ml models that the patient has skin cancer that are accurate 96% of times \n '
    for i in x:
      ou+=f"{i['label']} : {i['score']}\n "
    ou+="prediction ends here\n"
    f= open('results.yml','a')
    f.write(ou)
    f.close()
    print(ou)
    return x
```

```python
def retinal_fundus_predi(path_image):
    from PIL import Image
    img_pil = Image.fromarray((path_image * 255).astype(np.uint8))

    from transformers import pipeline
    pipe = pipeline("image-classification", model="pamixsun/swinv2_tiny_for_glaucoma_classification")
    x = pipe(img_pil)
    ou='Prediction of specialized ml models that the patient has glaucoma \n '
    for i in x:
      ou+=f"{i['label']} : {i['score']}\n "
    ou+="prediction ends here\n"
    f= open('results.yml','a')
    f.write(ou)
    f.close()
    print(ou)
    return x
```

```python
def pneumonia_xray_predi(path_image):
    from transformers import pipeline
    from PIL import Image
    img_pil = Image.fromarray((path_image * 255).astype(np.uint8))
    pipe = pipeline("image-classification", model="nickmuchi/vit-base-xray-pneumonia")
    x = pipe(img_pil)
    ou="Prediction label with confidence that are accurate to 90% \n "
    for i in x:
      ou+=f"{i['label']} : {i['score']}\n "
    ou+="prediction ends here\n"
    f= open('results.yml','a')
    f.write(ou)
    f.close()
    return x
```

```python
def wrapperfun(bolf,img):
  if bolf=="MRI":
    return brain_tumor_mri_predi(img), alzheimer_mri_predi(img)
  elif bolf=="Eye Scan":
    return retinal_fundus_predi(img)
  elif bolf=="Skin Image":
    return skin_cancer_predi(img)
  else:
    return pneumonia_xray_predi(img)
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')
chat = model.start_chat(history=[])
def generator(message,history1):

  f= open('patient.txt','r')
  history_p=f.read()
  f.close()
  f= open('results.yml','r')
  history_ana_p=f.read()
  f.close()

  prompt = '''Task:
Analyze the patient's summary and the specialized model predictions to provide a comprehensive recommendation. Consider the following points in your analysis:

1. Highlight any significant factors in the patient's history, including their current diseases and medications.
2. Summarize the model predictions, emphasizing the most likely conditions and their probabilities.
3. Suggest the best course of action for the healthcare provider, including:
   - Immediate steps for diagnosis and treatment.
   - Additional tests to confirm the model predictions.
   - Any potential interactions or concerns related to the patient's current medications and known conditions.
4. Provide a detailed and clear recommendation that addresses the patient's current condition and potential risks based on the model predictions.

Your analysis and recommendation should be concise, actionable, and based on the provided patient details and model predictions.'''
  if not history1:
    response = chat.send_message("Patient summary "+history_p+"predictions by specialized models "+history_ana_p+prompt)
  else:
    response = chat.send_message(message)
  f= open('chat_report.txt','a')
  f.write("User : "+message+ "\n Model : " +response.text +"\n")
  f.close()
  return response.text

```

```python
import gradio as gr
import numpy as np
import pandas as pd

# Medical form submission function
def submit_form(name, age, sex, contact_number, email_address, weight, height, diseases, blood_group, medication, list_medication, allergy, list_allergy,reason):
    formatted_data = f"Patient name - {name}, Age - {age}, sex - {sex}, weight - {weight}, height - {height}, blood group - {blood_group}"
    if diseases:
        formatted_data += ", diseases - " + ", ".join(diseases)
    else:
        formatted_data += ", diseases - None"
    if medication == "Yes":
        formatted_data += ", currently on medication - " + medication + ", list of current medication- " + list_medication
    else:
        formatted_data += ", currently on medication - " + medication
    if allergy == "Yes":
        formatted_data += ", known allergy - " + allergy + ", list of known allergies- " + list_allergy
    else:
        formatted_data += ", known allergy - " + allergy
    formatted_data +=", purpose of visit - "+reason
    file_name = f"patient.txt"
    f= open(file_name,'w')
    f.write(formatted_data)
    f.close()

    return formatted_data, gr.update(visible=True)

# Medical form interface
form_interface = gr.Interface(
    fn=submit_form,
    inputs=[
        gr.Textbox(label="Name"),
        gr.Number(label="What is your age?"),
        gr.Radio(choices=["Male", "Female", "Other"], label="Sex"),
        gr.Textbox(label="Contact Number"),
        gr.Textbox(label="Email Address"),
        gr.Number(label="Weight"),
        gr.Textbox(label="Height"),
        gr.CheckboxGroup(choices=["Diabetes", "Hypertension", "Cardiac Disease", "Cancer", "Asthma", "Psychiatic Disorder", "Epilepsy", "Other", "None"], label="Check the conditions that apply to you or any member of immediate relatives:"),
        gr.Dropdown(choices=["A+", "A-", "B+", "B-", "AB+", "AB-", "O+", "O-"], label="Blood Group"),
        gr.Radio(choices=["Yes", "No"], label="Are you currently taking any medication?"),
        gr.Textbox(label="Medication Details (separated by commas)"),
        gr.Radio(choices=["Yes", "No"], label="Do you have any allergies?"),
        gr.Textbox(label="List your Allergy Details (separated by commas)"),
        gr.Textbox(label="Reason for visit")
    ],
    outputs=[
        gr.Textbox(label="Submitted Data"),
        gr.Button(visible=False)
    ],
    title="Medical Form"
)

# Reports upload interface
upload_interface = gr.Interface(
    fn=wrapperfun,
    inputs=[
        gr.Radio(choices=["MRI", "Eye Scan", "Skin Image", "X-ray"], label="Choose your type of diagnostic image or scan:"),
        #gr.Image(label="X-Ray"),
        #gr.Image(label="MRI"),
        gr.Image(label="Scans")
    ],
    outputs=[
        #gr.Image(label="X-Ray Sepia"),
        #gr.Image(label="MRI Sepia"),
        gr.Textbox(label="Analysis")
    ],
    title="Upload Medical Reports"
)

def process_csv(file):
    df = pd.read_csv(file.name)
    return df.head()

# blood_test_interface = gr.Interface(
#     fn=process_csv,  # Function to process the uploaded file
#     inputs=gr.File(label="Upload CSV File"),  # File upload input
#     outputs=gr.Textbox(label="Analysis"),  # Output as a DataFrame
#     title="CSV File Uploader",  # Title of the interface
#     description="Upload a CSV file and see the first few rows of data."  # Description of the interface
# )
chatbot = gr.ChatInterface(fn=generator, examples=["initiate"], title="IMCI")

# Create a tabbed interface to switch between form and upload pages
demo = gr.TabbedInterface([form_interface, upload_interface, chatbot], ["Medical Form", "Upload Scans","Assistant"])

if __name__ == "__main__":
    demo.launch(debug=True)
```
